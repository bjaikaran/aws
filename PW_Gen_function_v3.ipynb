```python
import sys
import os
import termcolor as tc
import random as rnd
import numpy as np
import pandas as pd
```

```python
class PasswordGen:
    
    #----------------------------------#
    # chr() range definitions          #
    #__________________________________#
    # UPPER =  chr range 65 to 90      #
    # LOWER =  chr range 97 to 122     #
    # NUMBER = chr range 48 to 57      #
    #----------------------------------#
    
    import importlib.util
    import sys
    import os   
  
    PYTHON_PACKAGES = {"package": {"termcolor":"tc", "pandas":"pd", "numpy":"np","random":"rnd"}}
    for package in PYTHON_PACKAGES["package"]:
        PI = package
        PI_AS = PYTHON_PACKAGES["package"][PI]
        IMP_STR = str("import "+PI+" as "+PI_AS)
        PIP_STR = str("python -m pip install "+PI)
        SPEC = importlib.util.find_spec(package)
        if SPEC is None:
            os.system(PIP_STR)
            exec(IMP_STR)
    
    SPECIAL = ("!", "@", "#", "$", "_", "=", "<", ">", "%")
    PW_FOUR = [0,0,0,0]
    
    
    def __init__(self,n):
        #exc_type, exc_value, exc_traceback = sys.exc_info()
        self.n = n
        if n < 0:
            (tc.cprint("ValueError: Value of password length cannot be negative. User entered {:,.0f}".format(n),"red", attrs={"bold"}))
            sys.exit(ValueError)
        elif n >128:
            (tc.cprint("ValueError: Value must be greater than 0 and no more than 127 characters. User entered {:,.0f}".format(n),color="blue", attrs={"bold"}))
            sys.exit(ValueError)
        else:
            return 
        exit
        
    
    def __repr__(self):
        RAND_PW = self.n - 4
        PW_TAIL = [[0] * self.n for i in range(RAND_PW)]
        for i in range(3):
            char = rnd.randint(1, 3)
            if char == 1:
                PH = chr(rnd.randint(65,90))
            elif char == 2:
                PH = chr(rnd.randint(97,122))
            else:
                PH = chr(rnd.randint(48,58))
            self.PW_FOUR[i] = str(PH)
        PW_HEAD = str(self.PW_FOUR[0])+str(self.PW_FOUR[1])+str(self.PW_FOUR[2])+str(self.PW_FOUR[3])
       
        for i in range(RAND_PW):
            char = rnd.randint(1, 4)
            if char == 1:
                pw = chr(rnd.randint(65,90))
            elif char == 2:
                pw = chr(rnd.randint(97,122))
            elif char == 3:
                pw = chr(rnd.randint(48,58))
            else: 
                pw = rnd.choice(self.SPECIAL)
            PW_TAIL[i] = pw
            
        
        PSWD_GEN = PW_HEAD+"".join(PW_TAIL)
        return PSWD_GEN
    

    
```

```python
def pw_range(x,y):
    import random as rnd
    x = x
    y = y
    if x < y and y < 128 and x > 0 :
        return rnd.randint(x,y)
    else:
        return print("range does not meet criteria")

```

```python
passwords = [[0] * 1 for i in range(550)]
```

```python
for i in range (550):
    passwords[i]=(PasswordGen(pw_range(35,127)))
    
```
